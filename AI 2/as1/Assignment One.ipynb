Linear and Polynomial Regression 

```python
import csv
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

```python
# A
# Load the training data hw1xtr.dat and hw1ytr.dat into the memory and plot it on one graph. 
dfXTrain = pd.read_table("hw1xtr.dat", header = None)
dfYTrain = pd.read_table("hw1ytr.dat", header = None)
plt.plot(dfXTrain, dfYTrain,'o',color ="black")
plt.xlabel('XTrain')
plt.ylabel('YTrain')
plt.title("X vs Y Training Data")
```

```python
# Load the test data hw1xte.dat and hw1yte.dat into the memory and plot it on another graph.
dfXTest = pd.read_table("hw1xte.dat", header = None)
dfYTest = pd.read_table("hw1yte.dat", header = None)
plt.plot(dfXTest, dfYTest,'o',color ="black")
plt.xlabel('XTest')
plt.ylabel('YTest')
plt.title("X vs Y Test Data")
```

```python
# B
# Add a column vector of 1’s to the features
xTest = np.loadtxt("hw1xte.dat") # Features, Add vector of 1's
XTest = np.c_[np.ones(xTest.size), xTest]
yTest = np.loadtxt("hw1yte.dat")
xTrain = np.loadtxt("hw1xtr.dat") # Add vector of 1's
XTrain = np.c_[np.ones(xTrain.size), xTrain]
yTrain = np.loadtxt("hw1ytr.dat")

# Use the linear regression formula discussed in Lecture 3 to obtain a 2-dimensional weight vector
# W = (XtransposeX)^-1XTransposeY
XTrans = np.transpose(XTrain)
Xnew = np.linalg.inv((XTrans)@XTrain)
W = Xnew @ XTrans @ yTrain

xS = np.linspace(xTrain.min(), xTrain.max(), xTrain.size)
Xn = np.c_[np.ones(xS.size), xS]
yn = Xn @ W

# Plot both the linear regression line and the training data on the same graph
fig, ax = plt.subplots(dpi = 120)
ax.plot(xS, yn)
ax.plot(xTrain,yTrain,'o', color = 'black')
plt.xlabel('XTrain')
plt.ylabel('YTrain')
plt.title("Linear Regression of Training Data")

# Report the average error on the training set using err = 1/m * sum i=1 to m (m>xi − yi)^2
m = xTrain.size
err = np.subtract(XTrain @ W,yTrain)
errNew = (np.sum(np.square(err))) / m
print('Average Error:',errNew)
```

```python
# C
xS = np.linspace(xTest.min(), xTest.max(), xTest.size)
Xn = np.c_[np.ones(xS.size), xS]
yn = Xn @ W
# Plot both the regression line and the test data on the same graph
fig, ax = plt.subplots(dpi = 120)
ax.plot(xS, yn)
ax.plot(xTest,yTest,'o', color = 'black')
plt.xlabel('XTest')
plt.ylabel('YTest')
plt.title("Linear Regression of Test Data")
# Report the average error on the test set using Eq.
m = xTest.size
err = np.subtract(XTest @ W,yTest)
errNew = (np.sum(np.square(err))) / m
print('Average Error:',errNew)
```

```python
# D
# Implement the 2nd-order polynomial regression by adding new features x^2 to the inputs.
XTest2 = np.c_[XTest, np.square(xTest)]
XTrain2 = np.c_[XTrain,np.square(xTrain)]

# Repeat (b) and (c). 

#Train
XTrans = np.transpose(XTrain2)
Xnew = np.linalg.inv((XTrans)@XTrain2)
W = Xnew @ XTrans @ yTrain

xS = np.linspace(xTrain.min(), xTrain.max(), xTrain.size)
Xn = np.c_[np.ones(xS.size), xS, np.square(xS)]
yn = Xn @ W

# Plot both the linear regression line and the training data on the same graph
fig, ax = plt.subplots(dpi = 120)
ax.plot(xS, yn)
ax.plot(xTrain,yTrain,'o', color = 'black')
plt.xlabel('XTrain')
plt.ylabel('YTrain')
plt.title("2nd Order Linear Regression of Training Data")

# Report the average error on the training set using err = 1/m * sum i=1 to m (m>xi − yi)^2
m = xTrain.size
err = np.subtract(XTrain2 @ W,yTrain)
errNew = (np.sum(np.square(err))) / m
print('Average Error:',errNew)
```

```python
#Test 
xS = np.linspace(xTest.min(), xTest.max(), xTest.size)
Xn = np.c_[np.ones(xS.size), xS, np.square(xS)]
yn = Xn @ W

# Plot both the linear regression line and the training data on the same graph
fig, ax = plt.subplots(dpi = 120)
ax.plot(xS, yn)
ax.plot(xTest,yTest,'o', color = 'black')
plt.xlabel('XTest')
plt.ylabel('YTest')
plt.title("2nd Order Linear Regression of Test Data")

# Report the average error on the training set using err = 1/m * sum i=1 to m (m>xi − yi)^2
m = xTest.size
err = np.subtract(XTest2 @ W,yTest)
errNew = (np.sum(np.square(err))) / m
print('Average Error:',errNew)
```

Is it a better fit than linear regression?

Yes as it has a better represenation of the data.

```python
# E
# Implement the 3rd-order polynomial regression by adding new features x^2, x3 to the inputs.
XTest3 = np.c_[XTest, np.square(xTest), np.power(xTest,3)]
XTrain3 = np.c_[XTrain,np.square(xTrain), np.power(xTrain,3)]

# Repeat (b) and (c).
#Train
XTrans = np.transpose(XTrain3)
Xnew = np.linalg.inv((XTrans)@XTrain3)
W = Xnew @ XTrans @ yTrain

xS = np.linspace(xTrain.min(), xTrain.max(), xTrain.size)
Xn = np.c_[np.ones(xS.size), xS, np.square(xS), np.power(xS,3)]
yn = Xn @ W

# Plot both the linear regression line and the training data on the same graph
fig, ax = plt.subplots(dpi = 120)
ax.plot(xS, yn)
ax.plot(xTrain,yTrain,'o', color = 'black')
plt.xlabel('XTrain')
plt.ylabel('YTrain')
plt.title("3rd Order Linear Regression of Training Data")

# Report the average error on the training set using err = 1/m * sum i=1 to m (m>xi − yi)^2
m = xTrain.size
err = np.subtract(XTrain3 @ W,yTrain)
errNew = (np.sum(np.square(err))) / m
print('Average Error:',errNew)
```

```python
#Test
xS = np.linspace(xTest.min(), xTest.max(), xTest.size)
Xn = np.c_[np.ones(xS.size), xS, np.square(xS), np.power(xS,3)]
yn = Xn @ W

# Plot both the linear regression line and the training data on the same graph
fig, ax = plt.subplots(dpi = 120)
ax.plot(xS, yn)
ax.plot(xTest,yTest,'o', color = 'black')
plt.xlabel('XTest')
plt.ylabel('YTest')
plt.title("3rd Order Linear Regression of Training Data")

# Report the average error on the training set using err = 1/m * sum i=1 to m (m>xi − yi)^2
m = xTest.size
err = np.subtract(XTest3 @ W,yTest)
errNew = (np.sum(np.square(err))) / m
print('Average Error:',errNew)
```

```python
# F
# Implement the 4th-order polynomial regression by adding new features x^2, x^3, x^4 to the inputs.
XTest4 = np.c_[XTest, np.square(xTest), np.power(xTest,3), np.power(xTest,4)]
XTrain4 = np.c_[XTrain,np.square(xTrain), np.power(xTrain,3), np.power(xTrain,4)]
# Repeat (b) and (c). 
#Train
XTrans = np.transpose(XTrain4)
Xnew = np.linalg.inv((XTrans)@XTrain4)
W = Xnew @ XTrans @ yTrain

xS = np.linspace(xTrain.min(), xTrain.max(), xTrain.size)
XnTrain = np.c_[np.ones(xS.size), xS, np.square(xS), np.power(xS,3), np.power(xS,4)]
ynTrain = XnTrain @ W

# Plot both the linear regression line and the training data on the same graph
fig, ax = plt.subplots(dpi = 120)
ax.plot(xS, ynTrain)
ax.plot(xTrain,yTrain,'o', color = 'black')
plt.xlabel('XTrain')
plt.ylabel('YTrain')
plt.title("4th Order Linear Regression of Training Data")

# Report the average error on the training set using err = 1/m * sum i=1 to m (m>xi − yi)^2
m = xTrain.size
err = np.subtract(XTrain4 @ W,yTrain)
errNew = (np.sum(np.square(err))) / m
print('Average Error:',errNew)
```

```python
# A
# Using the training data to implement l2-regularized for the 4th-order polynomial regression 
# (page 12 of Lecture 4, note that we do not penalize the bias term w0), vary the regularization parameter 
# λ ∈ {0.01, 0.1, 1, 10, 100, 1000, 10000}. 

l = np.array([0.01, 0.1, 1, 10, 100, 1000, 10000])
error = np.array([])
errorT = np.array([])
wFinal = []

Xtrans = np.transpose(XTrain4)
identity = np.identity(5)
identity[0][0] = 0
print(identity)
m = 40

for i in l:
    # W = ((Xtranspose)(X) + λI^)inverse XTranpose(y)
    W = (np.linalg.inv(((Xtrans)@XTrain4)+ (i * identity))) @ XTrans @yTrain
    print(W)
    xnTrain = XTrain4 @ W
    errS = np.subtract(XTrain4 @ W, yTrain)
    err = (np.sum(np.square(errS)))/m
    error = np.append(error,err)
    errST = np.subtract(XTest4 @ W, yTest)
    errT = (np.sum(np.square(errST)))/xTest.size
    errorT = np.append(errorT,errT)
    wFinal.append(W)
    
# Plot the training and test error (averaged overall instances) using Eq. (1) as a function of λ (you should use a log10 scale for λ).
fig, ax = plt.subplots(dpi = 120)
ax.loglog(l, error, color = "red")
ax.loglog(l, error,'o' ,color = "blue")
ax.loglog(l, errorT, color = "black")
ax.loglog(l, errorT,'o' ,color = "yellow")
plt.xlabel('XTrain')
plt.ylabel('YTrain')
plt.title("4th Order L2 Regularization of Training Data")
```

Which λ is the best for fitting the data?

The best lambda is 0.01

```python
# B
# Plot the value of each weight parameter (including the bias term w0) as a function of λ.

split0 = np.array([item[0] for item in wFinal])
split1 = np.array([item[1] for item in wFinal])
split2 = np.array([item[2] for item in wFinal])
split3 = np.array([item[3] for item in wFinal])
split4 = np.array([item[4] for item in wFinal])

fig, ax = plt.subplots(dpi = 120)
ax.plot(split0, color ='red')
ax.plot(split1, color ='blue')
ax.plot(split2, color ='black')
ax.plot(split3, color ='yellow')
ax.plot(split4, color ='orange')
plt.xlabel('Lambda')
plt.ylabel('Weighted Parameters')
plt.title("Lambda's vs Weighted Parameters")
```

```python
# C 
# Write a procedure that performs five-fold cross-validation on your training data (page7 of Lecture 4). 
xFolds = np.split(XTrain4, 5)
yFolds = np.split(yTrain,5)
print(yTrain)
fold = 0
errors=[]

for i in l:
    print('lambda value: ',i,'\n')
    fold = 0
    error = 0
    for f in xFolds:
        trainingSetx = np.delete(xFolds, fold, 0)
        validationSetx = xFolds[fold]
        trainingSetx = np.concatenate(trainingSetx)
        trainingSety = np.delete(yFolds, fold, 0)
        validationSety = yFolds[fold]
        trainingSety = np.concatenate(trainingSety)
        w = np.linalg.inv((trainingSetx.T@trainingSetx) + (i*identity)) @ trainingSetx.T @ trainingSety
        m = xTrain.size
        errt = np.subtract(validationSetx@w,validationSety)
        err = (np.sum(np.square(errt))) / m
        error += err
        fold += 1
        
#Show the average error on the validation set as a function of λ
    error = error/5
    print('avg error is: ', error)
    errors.append(error)
    
fig, ax = plt.subplots(dpi = 120)
ax.loglog(l, errors, color = "red")
ax.loglog(l, errors,'o' ,color = "blue")
plt.xlabel('Regularization Parameters')
plt.ylabel('Average Val Error')
plt.title("Average Val Error vs Regularization Parameters")
```
